```python
lis= []

with open("TCS Technical Analysis (TCS) - Investing.com.html", encoding="utf-8") as f:
    data = f.read()
    soup = BeautifulSoup(data, 'html.parser')
    company =  soup.find("div", {"class":"wrapper"})
    f.close()

company = company.find("section", {"id":"leftColumn"})
name_of_the_company = company.find("div", {"class":"instrumentHeader"})
name_of_the_company = name_of_the_company.find("h2", {"class":"float_lang_base_1 inlineblock"}).text

first = []
second = []

summary = company.find("div", {"id":"techinalContent"})
summary = summary.find("div", {"class":"newTechStudiesRight instrumentTechTab"})#.text
summary = summary.findAll("div", {"class":"summaryTableLine"})

for i in range(0,len(summary)):
    if i==0:
        name = summary[i].findAll("span")[0].text
        first.append(name)
        
        table = summary[i].findAll("span")[1].text
        first.append(table.title())     

        row1 = summary[i].findAll("span")[2]
        row1_1 = row1.find("i", {"class":"noBold"}).text
        row1_2 = row1.find("i", {"id":"maBuy"}).text
        row_1 = row1_1 + row1_2
        first.append(row_1)

        row2 = summary[i].findAll("span")[3]
        row2_1 = row2.find("i", {"class":"noBold"}).text
        row2_2 = row2.find("i", {"id":"maSell"}).text
        row_2 = row2_1 + row2_2
        first.append(row_2)
    
    if i==1:
        name = summary[i].findAll("span")[0].text
        second.append(name)
        
        table = summary[i].findAll("span")[1].text
        second.append(table.title())
        
        row1 = summary[i].findAll("span")[2]
        row1_1 = row1.find("i", {"class":"noBold"}).text
        row1_2 = row1.find("i", {"id":"tiBuy"}).text
        row_1 = row1_1 + row1_2
        second.append(row_1)
        
        row2 = summary[i].findAll("span")[3]
        row2_1 = row2.find("i", {"class":"noBold"}).text
        row2_2 = row2.find("i", {"id":"tiSell"}).text
        row_2 = row2_1 + row2_2
        second.append(row_2)

lis.append(first)
lis.append(second)

data = pd.DataFrame(lis, columns = [name_of_the_company, None, None, None])#, index=data[0])
data = data.set_index(data.columns[0])

print(data)
data.to_csv(name_of_the_company+".csv")
```
